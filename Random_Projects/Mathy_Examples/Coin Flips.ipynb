```python
import random
import numpy as np
import matplotlib.pyplot as plt
print(random.random())
```

```python
# 1 represents Heads, 0 represents tails
int('11111',2)
```

```python
num_tosses = int(100000)
num_coins = 10
results_tracker = np.zeros((2**num_coins, 1), dtype="int64")
for _ in range(num_tosses):
    cur_flip = ''
    tosses = np.random.random(size = num_coins)
    for i in range(num_coins):
        if tosses[i] >= 0.5:
            cur_flip += str(1)
        else:
            cur_flip += str(0)
    dec_vers = int(cur_flip, 2)
    results_tracker[dec_vers] += 1
plt.scatter(range(2**num_coins),results_tracker)
plt.hlines(y = num_tosses/(2**num_coins), xmin = 0, xmax = 2**num_coins)
# plt.ylim(0, max(results_tracker))
plt.xlim(0, 2**num_coins)
plt.ylabel("Occurences")
plt.xlabel("Decimal Representation of Binary")
plt.title("Each outcome is equally likely")
plt.show()
```

```python
num_tosses = int(100000)
num_coins = 50
results_tracker = np.zeros((num_coins + 1, 1), dtype="int32")
for _ in range(num_tosses):
    cur_flip = 0
    tosses = np.random.random(size = num_coins)
    for i in range(num_coins):
        if tosses[i] >= 0.5:
            cur_flip += 1
        else:
            cur_flip += 0
    results_tracker[cur_flip] += 1
plt.scatter(range(num_coins + 1),results_tracker)
plt.hlines(y = num_tosses/(num_coins), xmin = 0, xmax = 2**num_coins)
# plt.ylim(0, max(results_tracker))
plt.xlim(0, num_coins)
plt.xlabel("Number of Heads")
plt.ylabel("Number of Occurences")
plt.title("The number of heads is not equally likely")
plt.show()
```
